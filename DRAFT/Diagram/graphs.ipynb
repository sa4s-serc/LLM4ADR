```python
import numpy as np

import matplotlib.pyplot as plt

categories = ['Flan-T5-base', 'Llama-3-8b-it', ' Gemma-2-9b-it', 'GPT-4o', 'Gemini-1.5 pro']
subcategories = ['Baseline', 'RAG', 'Finetune', 'DRAFT']
data = np.random.rand(5, 4)  # Random data for demonstration

# Data
fig, ax = plt.subplots()
F1 = [[0.734, 0.772, 0.855, 0.885],
    [0.805, 0.807, 0.837, 0.848],
    [0.810, 0.807, 0.847, 0.842],
    [0.814, 0.836, 0, 0],
    [0.817, 0.831, 0, 0]]


bar_width = 0.2
index = np.arange(len(categories))

colors = ['red', 'blue', 'orange', 'green']
for i in range(len(subcategories)):
    ax.bar(index + i * bar_width, [row[i] for row in F1 if len(row) > i], bar_width, label=subcategories[i], color=colors[i])
    
plt.xticks(rotation=15)
ax.set_ylim(0.7, 0.9)
ax.set_xlabel('LLM')
ax.set_ylabel('Bert Score F1')
# ax.set_title('Bert Score F1 of various approaches')
ax.set_xticks(index + bar_width * 1.5)
ax.set_xticklabels(categories)
ax.legend()

plt.savefig('bar_graph_F1.png', dpi=1000)  # 300 DPI for higher resolution
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Categories and subcategories
categories = ['Flan-T5-base', 'Llama-3-8b-it', 'Gemma-2-9b-it', 'GPT-4o', 'Gemini-1.5 pro']
subcategories = ['Baseline', 'RAG', 'Finetune', 'DRAFT']

# Data
F1 = [[0.734, 0.772, 0.855, 0.885],
    [0.805, 0.807, 0.837, 0.848],
    [0.810, 0.807, 0.847, 0.842],
    [0.814, 0.836, 0, 0],
    [0.817, 0.831, 0, 0]]

bar_width = 0.15
index = np.arange(len(categories))

# Create figure
fig = go.Figure()

# Add bars for each subcategory
colors = ['red', 'blue', 'orange', 'green']
for i in range(len(subcategories)):
    fig.add_trace(go.Bar(
        x=[category for category in categories],  # x-axis categories
        y=[row[i] for row in F1],  # y-axis data for the subcategory
        name=subcategories[i],  # Name for the legend
        width=bar_width,  # Bar width
        marker_color=colors[i],  # Color of the bars
        offsetgroup=i  # This groups the bars together for each subcategory
    ))

# Customize layout
fig.update_layout(
    xaxis=dict(
        tickmode='array', 
        tickvals=index, 
        ticktext=categories,  # Set x-axis categories
        tickangle=-45,  # Rotate x-axis labels at 45 degrees
        title_font=dict(size=32),  # Larger x-axis title
        tickfont=dict(size=32)      # Larger x-axis tick labels
    ),
    yaxis=dict(
        range=[0.7, 0.9], 
        title="Bert Score F1",
        title_font=dict(size=32),  # Larger y-axis title
        tickfont=dict(size=32)      # Larger y-axis tick labels
    ),    barmode='group',  # Group bars together
    # title="Bert Score F1 of various approaches",  # Title of the chart
    xaxis_title="LLM",  # x-axis label
    # legend_title="DRAFT",  # Legend title
    # template='plotly_dark',  # Optional: Dark theme for better visibility
    autosize=False,  # Allow custom size
    width=1200,  # Increase the figure width (for higher resolution)
    height=800,  # Increase the figure height (for higher resolution)

    legend=dict(
        x=0.85,  # Set x position inside the plot area (0 to 1 scale)
        y=0.95,  # Set y position inside the plot area (0 to 1 scale)
        xanchor='left',  # Align legend to the left
        yanchor='top',  # Align legend to the top
        # bgcolor='rgba(255, 255, 255, 0.7)',  # Background color with transparency
        bordercolor='Black',  # Border color around the legend
        borderwidth=1,  # Border width around the legend
        font=dict(size=18)
    )
)

fig.show()

fig.write_image("plot.png", width=2000, height=1500, scale=2)  # scale=2 for higher resolution
```

```python
import pandas as pd
import matplotlib.pyplot as plt

approach_data = {
    'Flan-T5': pd.read_csv('../../Graphs/Approach/t5-lora/val.csv'),
    'Flan-T5 (Full)': pd.read_csv('../../Graphs/Approach/t5-full/val.csv'),
    'Gemma-2-9b-it': pd.read_csv('../../Graphs/Approach/Gemma/val.csv'),
    'Llama-3-8b-it ': pd.read_csv('../../Graphs/Approach/Llama/val.csv'),
}

finetuning_data = {
    'Flan-T5': pd.read_csv('../../Graphs/Finetuning/t5-lora/val.csv'),
    'Flan-T5 (Full)': pd.read_csv('../../Graphs/Finetuning/t5-full/val.csv'),
    'Gemma-2-9b-it': pd.read_csv('../../Graphs/Finetuning/Gemma/val.csv'),
    'Llama-3-8b-it ': pd.read_csv('../../Graphs/Finetuning/Llama/val.csv'),
}
```
