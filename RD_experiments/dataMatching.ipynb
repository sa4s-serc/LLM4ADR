```python
import pandas as pd
import numpy as np
import os
from pprint import pprint

csv_path = os.path.join("..", "0_shot", "results", "babbage-002.csv")
json_path = os.path.join("..", "Data", "ADR-Data", "data.jsonl")

json_data = pd.read_json('../Data/ADR-Data/data.jsonl', lines=True)
csv_data = pd.read_csv(csv_path)
json_data['Predictions'] = ''
```

```python
count = 0
Predictions = 'babbage-002'

for i, row in json_data.iterrows():
    for j, csv_row in csv_data.iterrows():
        if(row['Context'] in csv_row['Context']):
            json_data.at[i, ['Predictions']] = csv_row[Predictions]
            count += 1
            if(count%100 == 0):
                print(count)
            break
            
        
    
```

```python
path_to_json = os.path.join("..", "0_shot", "results", Predictions+".jsonl")
json_data.to_json(path_to_json, orient="records", lines=True)
```
