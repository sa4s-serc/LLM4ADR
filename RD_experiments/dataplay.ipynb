```python
import pandas as pd
import tiktoken
```

```python
num_tokens = 50
df_1 = pd.read_json("../Data/ADR-data/data.jsonl", lines=True)
encoding = tiktoken.get_encoding("cl100k_base")
df = df_1
df["Context_tokens"] = df["Context"].apply(lambda x: len(encoding.encode(x)))
df["Decision_tokens"] = df["Decision"].apply(lambda x: len(encoding.encode(x)))
df = df[df["Context_tokens"] <= num_tokens]
df = df[df["Decision_tokens"] <= num_tokens]
```

```python
len(df)
df.to_json("data/less_than_50.jsonl", orient="records", lines=True)
```

```python
# find unique contexts with frequencies
context_freq = df["Context"].value_counts().reset_index()
context_freq.columns = ["Context", "Frequency"]

# print all frequencies with more than 1
context_freq = context_freq[context_freq["Frequency"] > 2]
context_freq
```

```python
# find all rows with unique context and decision pairs
df_unique = df_1.drop_duplicates(subset=["Context", "Decision"])
len(df_1) - len(df_unique)
```
