```python
import pandas as pd
import matplotlib.pyplot as plt

approach_data = {
    'Flan-T5': pd.read_csv('../../Graphs/Approach/t5-lora/val.csv'),
    'Flan-T5 (Full)': pd.read_csv('../../Graphs/Approach/t5-full/val.csv'),
    'Gemma-2-9b-it': pd.read_csv('../../Graphs/Approach/Gemma/val.csv'),
    'Llama-3-8b-it ': pd.read_csv('../../Graphs/Approach/Llama/val.csv'),
}

finetuning_data = {
    'Flan-T5': pd.read_csv('../../Graphs/Finetuning/t5-lora/val.csv'),
    'Flan-T5 (Full)': pd.read_csv('../../Graphs/Finetuning/t5-full/val.csv'),
    'Gemma-2-9b-it': pd.read_csv('../../Graphs/Finetuning/Gemma/val.csv'),
    'Llama-3-8b-it ': pd.read_csv('../../Graphs/Finetuning/Llama/val.csv'),
}
```

```python
# for j, (model, df) in enumerate(approach_data.items()):
#     plt.plot(df['epoch'][:5], df['loss'][:5], label=model)
#     plt.xlabel('Epoch')
#     plt.ylabel('Loss')
#     # plt.title('Approach Validation Loss')
#     plt.legend()

# plt.savefig('../../Graphs/approach_loss.png', dpi=600)

# Use plotly to make the same graph
import plotly.express as px

# Combine data into a single DataFrame
data = pd.concat([
    pd.DataFrame({
        'epoch': approach_data['Flan-T5']['epoch'][:6],
        'loss': approach_data['Flan-T5']['loss'][:6],
        'model': 'Flan-T5'
    }),
    pd.DataFrame({
        'epoch': approach_data['Flan-T5 (Full)']['epoch'][:6],
        'loss': approach_data['Flan-T5 (Full)']['loss'][:6],
        'model': 'Flan-T5 (Full)'
    }),
    pd.DataFrame({
        'epoch': approach_data['Gemma-2-9b-it']['epoch'][:6],
        'loss': approach_data['Gemma-2-9b-it']['loss'][:6],
        'model': 'Gemma-2-9b-it'
    }),
    pd.DataFrame({
        'epoch': approach_data['Llama-3-8b-it ']['epoch'][:6],
        'loss': approach_data['Llama-3-8b-it ']['loss'][:6],
        'model': 'Llama-3-8b-it'
    })
])

# Plot using plotly.express
fig = px.line(
    data,
    x='epoch',
    y='loss',
    color='model',  # Group by model column
    labels={'epoch': 'Epoch', 'loss': 'Loss'}
)

fig.update_layout(
    legend_title_text='', 
    autosize=False,
    width=700,
    height=500,
    margin=dict(l=0, r=0, t=50, b=50),
    template='seaborn'
)

fig.show()
#  save the plot as a png file
fig.write_image('../../Graphs/approach_loss.png', width=700, height=500, scale=5)
```
